```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./walmart-sales-dataset-of-45stores.csv',parse_dates=['Date'])
data.head()
```

```python
sns.boxplot(data=data,y='Weekly_Sales')
plt.show()
```

```python
data['Z-score'] = stats.zscore(data.Weekly_Sales)
data.head()
```

```python
cleaned_data = data[data['Z-score']<3]
cleaned_data
```

```python
sns.boxplot(data=cleaned_data,y='Weekly_Sales')
plt.show()
```
